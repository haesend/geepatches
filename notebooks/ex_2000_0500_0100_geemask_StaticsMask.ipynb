```python
import geemap
import ee
if not ee.data._credentials: ee.Initialize()

import sys
sys.path.append('../src/geepatches')
import geeutils
import geeproduct
import geemask
```

```python

eepoint       = geeutils.antwerppoint
eedatefrom    = ee.Date('2018-01-01') 
eedatetill    = eedatefrom.advance(1, 'year')
eeroi         = geeutils.squareareaboundsroi(eepoint, 1000)
verbose       = False

s2sclclassesarray = [2,4,5,6,7]
threshold         = -30
thresholdunits    = "percentage"
fractionsmaker    = geemask.ClassFractions(s2sclclassesarray)
staticsmaker      = geemask.StaticsMask(s2sclclassesarray, None, threshold, thresholdunits, None, True)

eeimagecollection = geeproduct.GEECol_s2scl().collect(eeroi, eedatefrom, eedatetill, verbose=verbose)
fractionsimage    = fractionsmaker.makefractions(eeimagecollection)
staticsimage      = staticsmaker.makemask(eeimagecollection)


xin = fractionsimage.clip(eeroi).reduceRegion(ee.Reducer.min()).values().get(0).getInfo()
xax = fractionsimage.clip(eeroi).reduceRegion(ee.Reducer.max()).values().get(0).getInfo()
xav = fractionsimage.clip(eeroi).reduceRegion(ee.Reducer.mean()).values().get(0).getInfo()
print (f"fractionsimage min({xin}) max({xax}) avg({xav})")


map = geemap.Map(height='600px')
map.centerObject(eepoint, 11)
map.addLayer(fractionsimage,                              {'min':xin,  'max':xax},   'fractionsimage') 
map.addLayer(fractionsimage.lte(abs(threshold)/100),      {'min':0,    'max':1  },   'lte')
map.addLayer(staticsimage ,                               {'min':0,    'max':1  },   'statics')                        
map.addLayer(geeutils.outlinegeometryimage(eeroi, 0, 1),  {'palette':'#ff0000'})
map
```
